```python
#Import Libraries
import pandas as pd 
import numpy as np
```

```python
# Sneak Peak of Data
df = pd.read_csv('lab2.csv')
df.head(5)
```

```python
# Check data shape
print(df.info)
```

```python
# Convert data into proper format & print first 5 transaction
transaction = []
temp_arr = []
next_acc = df.values[0,0]
count = 0
for i in range(0,df.shape[0]):
    last_acc = next_acc
    next_acc = df.values[i,0]
    if (last_acc == next_acc):
        temp_arr.append(str(df.values[i,1]))
    else:
        transaction.append(temp_arr)
        temp_arr = [df.values[i,1]]
    if (i == 32366):
        transaction.append(temp_arr)

print(transaction[0:5])
```

```python
# Perform Apriori and disply first 5 result
from apyori import apriori
rule = apriori(transactions = transaction, min_support = 0.05)
result = list(rule)

print(result[0:5])
```

```python
# Convert Result into dataframe
format_result = []

for i in result:
    for j in range(0,len(i[2])):
        item_base = tuple(i[2][j][0])
        item_add = tuple(i[2][j][1])
        support = i[1]
        confidance = i[2][j][2]
        lift = i[2][j][3]
        format_result.append([item_base, item_add,support, confidance,lift])
        
result_df = pd.DataFrame(format_result,columns=["Base Item","Add Item","Support", "Confidance","Lift"])
print(result_df)
```

```python
# Sort Result by Lift
sort_df = result_df.sort_values(by=['Lift'],ascending=False)
print(sort_df)
```
